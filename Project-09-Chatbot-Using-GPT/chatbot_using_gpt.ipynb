```python
import openai
```

```python
openai.api_key = os.getenv("OPENAI_API_KEY")
```

```python
def virtual_assistant(message, chat_history = []):
    # Define Intial User Message
    user_message = message
    # Create a List of Messages
    messages = [{'role':'system', 'content':'Act as Travel Assistant'}]
    messages.extend([{'role':'user', 'content':msg} for msg in chat_history])
    messages.append({'role':'user','content':user_message})

    # Generate a response from the model
    response = openai.ChatCompletion.create(
        model = 'gpt-3.5-turbo',
        messages = messages,
        temperature = 0.7,           # Controls randomness in the output (0.0 to 1.0)
        max_tokens = 400             # Maximum number of tokens to generate
    )

    model_reply = response['choices'][0]['message']['content']
    print("Travel Assistant : ", model_reply)
    return model_reply
```

```python
# Start a conversation with chatgpt
chat_history = []
print("Trip Advisor: Hello! How can I assist you today?")
while True:
    user_input = input("User: ")
    if user_input.lower() == "quit" or user_input.lower() == "thanks":
        break
    chatbot_reply = virtual_assistant(user_input, chat_history=[])
    chat_history.append(user_input)
    chat_history.append(chatbot_reply)
```
